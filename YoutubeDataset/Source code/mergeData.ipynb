```python
import glob
import pandas as pd
import re
```

```python
path = "/content/"
 

file_list = glob.glob(path + "/*.xlsx")
 
excl_list = []
 
for file in file_list:
    excl_list.append(pd.read_excel(file))

excl_merged = pd.concat(excl_list, ignore_index=True)   



# loai ID trung
excl_merged = excl_merged.sort_values("videoId")
excl_merged.drop_duplicates(subset ="videoId", keep = False, inplace = True)
excl_merged = excl_merged.sample(frac=1).reset_index(drop=True)

excl_merged = excl_merged.fillna(value={"Duration":"PT0S"})
```

```python
excl_merged.Duration = excl_merged.Duration.map(lambda x: str(isodate.parse_duration(x)))
excl_merged['viewCount'] = pd.to_numeric(excl_merged['viewCount'], downcast='integer')
excl_merged['subscriberCount'] = pd.to_numeric(excl_merged['subscriberCount'], downcast='integer')
excl_merged['viewCount.1'] = pd.to_numeric(excl_merged['viewCount.1'], downcast='integer')
excl_merged['likeCount'] = pd.to_numeric(excl_merged['likeCount'], downcast='integer')
excl_merged.head()
```

```python
excl_merged.to_excel('/content/drive/MyDrive/DoAnTienXuLiDuLieu/Dataset/merged_data.xlsx', index=False)
```
