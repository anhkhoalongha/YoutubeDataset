```python
import pandas as pd 
data=pd.read_csv('drive/MyDrive/DoDongThuan2.csv')
data
```

```python
col1= data['annotator1']
col2= data['annotator2']
```

```python
#Autos & Vehicles -> 0
#Comedy ->1
#Education ->2
#Entertainment  ->3
#Film and Animation ->4
#Gaming  -> 5
#Howto & Style -> 6
#Music  -> 7
#News & Politics  -> 8
#Nonprofits & Activism -> 9
#People & Blogs -> 10
#Pets & Animals -> 11
#Science & Technology ->12
#Sports ->13
#Travel & Events  ->14
col1.replace({"Autos & Vehicles": 0, "Comedy": 1, "Education": 2, "Entertainment": 3, "Film and Animation": 4, "Gaming": 5, "Howto & Style": 6, "Music": 7, "News & Politics": 8, "Nonprofits & Activism": 9, "People & Blogs": 10, "Pets & Animals": 11, "Science & Technology": 12, "Sports": 13, "Travel & Events": 14}, inplace=True)
col2.replace({"Autos & Vehicles": 0, "Comedy": 1, "Education": 2, "Entertainment": 3, "Film and Animation": 4, "Gaming": 5, "Howto & Style": 6, "Music": 7, "News & Politics": 8, "Nonprofits & Activism": 9, "People & Blogs": 10, "Pets & Animals": 11, "Science & Technology": 12, "Sports": 13, "Travel & Events": 14}, inplace=True)
list_label =["Autos & Vehicles", "Comedy", "Education", "Entertainment", "Film and Animation", "Gaming", "Howto & Style", "Music", "News & Politics", "Nonprofits & Activism", "People & Blogs", "Pets & Animals", "Science & Technology", "Sports", "Travel & Events"]
```

```python
import numpy as np
A = np.zeros( (16, 16) )
i=0
while(i<299):
  if (col1[i]==col2[i]):
    A[col1[i]][col2[i]]=A[col1[i]][col2[i]]+1
  i=i+1
A
```

Xác xuất đồng thuận 2 người:

```python
Pr_A = 0
for i in range(0,15):
  Pr_A= Pr_A + A[i][i]/299
Pr_A
```

Xác suất xảy ra đồng thuận giữa hai người đối với từng nhãn:

```python
pr1= np.zeros(16)
pr2= np.zeros(16)
pr=np.zeros(15)
for j in range(0,15):
  pr1[j]=0
  pr2[j]=0
  for i in range(0,15):
    pr1[j] = pr1[j] + A[j][i]/299
    pr2[j] = pr2[j] + A[i][j]/299
  pr[j] = pr1[j]*pr2[j]
for i in range(0, 15):
  print(list_label[i], ": ",pr[i])
```

Xác suất giả định của 2 annotator đồng thuận trên 15 nhãn

```python
Pr_e=0
for i in range(0,15):
  Pr_e= Pr_e+ pr[i]
Pr_e
```

Độ đồng thuận giữa hai người

```python
k=0
k=(Pr_A-Pr_e)/(1-Pr_e)
k
```
